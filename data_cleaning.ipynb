```python
import pandas as pd
import numpy as np
import codecs, json
```

```python
games = pd.read_csv("soccerteams-arenas.csv")
```

```python
games.head()
```

```python
data_counts=pd.DataFrame({'count':games.groupby(['Team 1', 'Team 2']).
                           size()}).reset_index()
```

```python
data_matrix=data_counts.pivot(index='Team 1', columns='Team 2', values='count')
```

```python
data_matrix=data_matrix.fillna(value=0)
```

```python
data_matrix
```

```python
##find index names that also match column names
matching_cols = [s for s in list(data_matrix.index.values) if s in list(data_matrix.columns.values)]
```

```python
select_matrix=data_matrix.loc[matching_cols][matching_cols]
select_matrix # nxn matrix
```

```python
p=np.array(select_matrix)
b=p.tolist()
json.dump(b, codecs.open('games_matrix.json', 'w', encoding='utf-8'), separators=(',', ':'), sort_keys=True, indent=4)
```

```python
countries=pd.DataFrame(select_matrix.index)
```

```python
countries.to_csv('country-colours.csv', index=False)##open the .csv file and insert HTML color codes next to name column.
```
